Question 1 - Extracting Tesla Stock Data Using yfinance - 2 Points


```python
import yfinance as yf
from tabulate import tabulate

# Download Tesla data for the maximum period
tsla = yf.Ticker("TSLA")
tesla_data = tsla.history(period="max")

# Check for "Date" column (optional)
# print(tesla_data.head())



# Select relevant columns
data_to_display = tesla_data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Display data in a table
print(tabulate(data_to_display, headers=data_to_display.columns, tablefmt="fancy_grid"))
```
